```python
# 0th: import modules
import pandas as pd
import os
import numpy as np
from tkinter import *
from tkinter import ttk,dialog
from tkinter import font
import json
import ast
import warnings
from datetime import datetime
warnings.filterwarnings("ignore")
## 1st: Check if database exists; if so load, if not create
data_columns = ['op id','Tarih','Ad','Soyad','Yaş','Cinsiyet','Cep Telefonu','TC Kimlik No','Adres','Yakınma','Fizik Muayene','Plan']
file_data=pd.DataFrame(columns=data_columns)
if (os.path.isfile('./data.csv')):
    file_data = pd.read_csv('./data.csv', index_col=[0])
file_data.to_csv('./data.csv')
file_data


city_file = open("cities.txt", "r", encoding='utf-8')
contents = city_file.read()
my_city_dict = ast.literal_eval(contents)
my_city_tuple= tuple(my_city_dict.keys())
city_file.close()
# 2nd: Define the functions that will manipulate the database
display_data = pd.DataFrame()
def refreshDisplay():
    global display_data
    display_data = file_data

refreshDisplay()
def addElement(element):
    data_added = pd.DataFrame(element, columns=data_columns)
    data_added.index.name='id'
    global file_data
    file_data = file_data.append(data_added,ignore_index=True)
    saveData(file_data)
    refreshDisplay()
    return file_data
    
def createElement(opid, date, name, last_name, age, sex, ph_number, id_no, location, complain, inspection, plan):
    element= ([[opid, str(date), str(name), str(last_name), str(age), str(sex), str(ph_number), str(id_no), str(location), (complain), (inspection), (plan)]])
    return element

def createTableElement(date, name, last_name, age, sex, ph_number, id_no, location, complain, inspection, plan):
    opid =0
    if(len(file_data['op id'])>0):
        opid = max(file_data['op id']) +1
    element= ([[opid, str(date), str(name), str(last_name), str(age), str(sex), str(ph_number), str(id_no), str(location), (complain), (inspection), (plan)]])
    return element

def removeElement(op_id):
    global file_data
    file_data = file_data[file_data["op id"] != op_id]
    saveData(file_data)
    refreshDisplay()
    return file_data
    
def editElement(op_id, element):
    global file_data
    file_data.loc[file_data['op id'] == op_id] = element;
    saveData(file_data)
    refreshDisplay()
    return file_data

def saveData(data):
    data.to_csv('./data.csv')  
    
def contains(element, array):
    l = []
    for i in array:
        l.append(element in str(i))
    return l
    
def searchElement(element, tree):
    if(element == "" or element == None):
        refreshDisplay()
        updateTree(tree)
    else:
        opid = element
        try:
            opid = int(element)
        except ValueError:
            pass
        
        condition = (opid == file_data['op id']) | contains(element, file_data['Tarih']) | contains(element, file_data['Ad']) | contains(element, file_data['Soyad']) | contains(element, file_data['Cep Telefonu']) | contains(element, file_data['TC Kimlik No']) | contains(element, file_data['Adres']) | contains(element, file_data['Yakınma']) | contains(element, file_data['Fizik Muayene']) |  contains(element, file_data['Plan']) | contains(element, file_data['Yaş']) | contains(element, file_data['Cinsiyet'])
        global display_data
        display_data = file_data[condition]
        updateTree(tree)
        # 3rd: Create gui and define basic operations

#update tree function
def updateTree(datatree):
    #clear tree
    for i in datatree.get_children():
        datatree.delete(i)
    #put dataframe data in tree
    data_rows = display_data.to_numpy().tolist()
    for row in data_rows:
        datatree.insert("","end",values=row)
        
    return


#delete selection function   
def deleteSelection(datatree, selected):
    op_id = datatree.item(selected)['values'][0]
    removeElement(op_id)
    #print(condition)
    datatree.delete(selected)
        

# 3.1: Create data addition form window
def patient_add_form_window(tree):
    add_patient = Tk()
    add_patient.resizable(height = False, width = False)
    add_patient.geometry("520x830")
    add_patient.title("Hasta Ekleme Formu")


    heading= Label(add_patient,text='Hasta Ekleme Formu', bg='grey', fg="white", width="500", height="3")
    heading.pack()




    date_text       = Label(add_patient, text='Tarih ')
    firstname_text  = Label(add_patient, text='İsim ')
    lastname_text   = Label(add_patient, text='Soyisim ')
    age_text        = Label(add_patient, text='Yaş ')
    sex_text        = Label(add_patient, text='Cinsiyeti')
    tel_text        = Label(add_patient, text='Cep Telefonu')
    tc_text         = Label(add_patient, text='TC No')
    adress_text     = Label(add_patient, text='Adres')
    complain_text   = Label(add_patient, text='Yakınma')
    inspection_text = Label(add_patient, text='Fizik Muayene')
    plan_text       = Label(add_patient, text='Plan')


    date_text.place(x=10, y=60)
    firstname_text.place(x=10, y=100)
    lastname_text.place(x=250, y=100)
    age_text.place(x=10, y=140)
    sex_text.place(x=250, y=140)
    tel_text.place(x=10, y=180)
    tc_text.place(x=250, y=180)
    adress_text.place(x=10, y=220)
    complain_text.place(x=10, y=320)
    inspection_text.place(x=10, y=500)
    plan_text.place(x=10, y=600)






    date       = StringVar()
    firstname  = StringVar()
    lastname   = StringVar()
    age        = StringVar()
    sex        = StringVar()
    tel        = StringVar()
    tc         = StringVar()
    adress_c   = StringVar()
    adress_t   = StringVar()
    adress_l   = StringVar()
    complain   = StringVar()
    inspection = StringVar()
    plan       = StringVar()


    date_entry       = Entry(add_patient, textvariable= date)
    firstname_entry  = Entry(add_patient, textvariable= firstname)
    lastname_entry   = Entry(add_patient, textvariable= lastname)
    age_entry        = Entry(add_patient, textvariable= age)
    sex_entry        = ttk.Combobox(add_patient, textvariable= sex, state='readonly')
    tel_entry        = Entry(add_patient, textvariable= tel)
    tc_entry         = Entry(add_patient, textvariable= tc)
    adress_entry_c   = ttk.Combobox(add_patient, textvariable= adress_c, state='readonly')
    adress_entry_t   = ttk.Combobox(add_patient, textvariable= adress_t, state='readonly')
    adress_entry_l   = ttk.Entry(add_patient, textvariable= adress_l)
    complain_entry   = Text(add_patient)
    inspection_entry = Text(add_patient)
    plan_entry       = Text(add_patient)
    
    sex_entry['values'] = ('Erkek',
                           'Kadın',
                           'Diğer',
                           'Belirtilmemiş')
    
    adress_entry_c['values']= my_city_tuple
    
    
    def citySelected(arg):
        adress_entry_t['values']= my_city_dict[adress_entry_c.get()]
        adress_entry_t.place(x=210, y=220, width="100", height="20")
    adress_entry_c.bind("<<ComboboxSelected>>", citySelected)
    

    date_entry.place(x=100, y=60, width="130", height="20")
    firstname_entry.place(x=100, y=100, width="130", height="20")
    lastname_entry.place(x=310, y=100, width="130",height="20")
    age_entry.place(x=100, y=140, width="130", height="20")
    sex_entry.place(x=310,y=140, width="130", height="20")
    tel_entry.place(x=100, y=180, width="130", height="20") 
    tc_entry.place(x=310, y=180, width="130", height="20")
    adress_entry_c.place(x=100, y=220, width="100", height="20")
    adress_entry_t.place(x=210, y=220, width="100", height="20")
    adress_entry_l.place(x=100, y=250, width="400", height="20")
    complain_entry.place(x=100, y=320, width="400", height="160")
    inspection_entry.place(x=100, y=500, width="400", height="80")
    plan_entry.place(x=100, y=600, width="400", height="60")
    


    
    

    ##REGISTER FUNCTION
    def register_patient():
        age=''
        tel=''
        tc=''
        try:
            date = date_entry.get()
            firstname = firstname_entry.get()
            lastname = lastname_entry.get()

            age = int(age_entry.get()) if (age_entry.get() != None and age_entry.get() != '') else ''
            tel = int(tel_entry.get()) if (tel_entry.get() != None and tel_entry.get() != '') else ''
            tc = int(tc_entry.get()) if (tc_entry.get() != None and tc_entry.get() != '') else ''

            sex = sex_entry.get()
            adress = adress_entry_l.get()+", "+ adress_entry_t.get()+", "+adress_entry_c.get()
            complain = complain_entry.get(1.0,END)
            inspection = inspection_entry.get(1.0,END)
            plan = plan_entry.get(1.0,END)
            
        except ValueError:
            error_text.config(text='Hata: Lütfen Yaş, TC Kimlik Numarası ve Telefon Numarası alanlarının\n numerik bir girdi olduğuna emin olunuz.\n Bir değer girmeyecekseniz lütfen boş bırakınız')
            return
        
        row=createTableElement(
        date,
        firstname,
        lastname,
        age,
        sex,
        tel,
        tc,
        adress,
        complain,
        inspection,
        plan
        )
        addElement(row)
        tree.insert("","end",values=row[0]);

        add_patient.destroy()

    reg_button= Button(add_patient, text="Hastayı Kaydet", command = register_patient, bg='orange',fg='white', width="30")
    reg_button.place(x="155", y="700", height="30")
    
    error_text = Label(add_patient, text='')
    error_text.place(x="80", y="750")
    
    add_patient.mainloop()
    
# 3.2: Create data editing form window
def patient_edit_form_window(tree, selected):
    edit_patient = Tk()
    edit_patient.resizable(height = False, width = False)
    edit_patient.geometry("520x830")
    edit_patient.title("Hasta Düzenleme Formu")


    heading= Label(edit_patient,text='Hasta Düzenleme Formu', bg='grey', fg="white", width="500", height="3")
    heading.pack()

    values= tree.item(selected)['values']




    date_text       = Label(edit_patient, text='Tarih ')
    firstname_text  = Label(edit_patient, text='İsim ')
    lastname_text   = Label(edit_patient, text='Soyisim ')
    age_text        = Label(edit_patient, text='Yaş ')
    sex_text        = Label(edit_patient, text='Cinsiyeti')
    tel_text        = Label(edit_patient, text='Cep Telefonu')
    tc_text         = Label(edit_patient, text='TC No')
    adress_text     = Label(edit_patient, text='Adres')
    complain_text   = Label(edit_patient, text='Yakınma')
    inspection_text = Label(edit_patient, text='Fizik Muayene')
    plan_text       = Label(edit_patient, text='Plan')


    date_text.place(x=10, y=60)
    firstname_text.place(x=10, y=100)
    lastname_text.place(x=250, y=100)
    age_text.place(x=10, y=140)
    sex_text.place(x=250, y=140)
    tel_text.place(x=10, y=180)
    tc_text.place(x=250, y=180)
    adress_text.place(x=10, y=220)
    complain_text.place(x=10, y=320)
    inspection_text.place(x=10, y=500)
    plan_text.place(x=10, y=600)






    date       = StringVar()
    firstname  = StringVar()
    lastname   = StringVar()
    age        = StringVar()
    sex        = StringVar()
    tel        = StringVar()
    tc         = StringVar()
    adress_c   = StringVar()
    adress_t   = StringVar()
    adress_l   = StringVar()
    complain   = StringVar()
    inspection = StringVar()
    plan       = StringVar()


    date_entry       = Entry(edit_patient, textvariable= date)
    firstname_entry  = Entry(edit_patient, textvariable= firstname)
    lastname_entry   = Entry(edit_patient, textvariable= lastname)
    age_entry        = Entry(edit_patient, textvariable= age)
    sex_entry        = ttk.Combobox(edit_patient, textvariable= sex, state='readonly')
    tel_entry        = Entry(edit_patient, textvariable= tel)
    tc_entry         = Entry(edit_patient, textvariable= tc)
    adress_entry_c   = ttk.Combobox(edit_patient, textvariable= adress_c, state='readonly')
    adress_entry_t   = ttk.Combobox(edit_patient, textvariable= adress_t, state='readonly')
    adress_entry_l   = ttk.Entry(edit_patient, textvariable= adress_l)
    complain_entry   = Text(edit_patient)
    inspection_entry = Text(edit_patient)
    plan_entry       = Text(edit_patient)
        
    sex_entry['values'] = ('Erkek',
                           'Kadın',
                           'Diğer',
                           'Belirtilmemiş')
    
    adress_entry_c['values']= my_city_tuple
    
    
    def citySelected(arg):
        adress_entry_t['values']= my_city_dict[adress_entry_c.get()]
        adress_entry_t.place(x=210, y=220, width="100", height="20")
    adress_entry_c.bind("<<ComboboxSelected>>", citySelected)
    

    t_adress = StringVar
    t_adress = values[8]
    t_adress_0 = t_adress[:(t_adress.index(','))]
    t_adress_1 = t_adress[(t_adress.index(',')+2):]
    t_adress_2 = t_adress_1[(t_adress_1.index(',')+2):]
    t_adress_1 = t_adress_1[:(t_adress_1.index(','))]
    date_entry.insert(0, values[1])
    firstname_entry.insert(0, values[2])
    lastname_entry.insert(0, values[3])
    age_entry.insert(0, values[4])
    sex_entry.insert(0, values[5])
    tel_entry.insert(0, values[6])
    tc_entry.insert(0, values[7])
    adress_entry_l.insert(0, t_adress[:t_adress.index(',')])
    adress_entry_c.current(my_city_tuple.index(t_adress_2))
    adress_entry_t['values']= my_city_dict[adress_entry_c.get()]
    adress_entry_t.current(my_city_dict[t_adress_2].index(t_adress_1))
    
    
    complain_entry.insert(1.0, values[9])
    inspection_entry.insert(1.0, values[10])
    plan_entry.insert(1.0, values[11])



    date_entry.place(x=100, y=60, width="130", height="20")
    firstname_entry.place(x=100, y=100, width="130", height="20")
    lastname_entry.place(x=310, y=100, width="130",height="20")
    age_entry.place(x=100, y=140, width="130", height="20")
    sex_entry.place(x=310,y=140, width="130", height="20")
    tel_entry.place(x=100, y=180, width="130", height="20")
    tc_entry.place(x=310, y=180, width="130", height="20")
    adress_entry_c.place(x=100, y=220, width="100", height="20")
    adress_entry_t.place(x=210, y=220, width="100", height="20")
    adress_entry_l.place(x=100, y=250, width="400", height="20")
    complain_entry.place(x=100, y=320, width="400", height="160")
    inspection_entry.place(x=100, y=500, width="400", height="80")
    plan_entry.place(x=100, y=600, width="400", height="60")
    
    

    
    
    
    
    
    
    


    def edit_patient_info():
        age=''
        tel=''
        tc=''
        try:
            date = date_entry.get()
            firstname = firstname_entry.get()
            lastname = lastname_entry.get()
            age = int(age_entry.get()) if (age_entry.get() != None and age_entry.get() != '') else ''
            sex = sex_entry.get()
            tel = int(tel_entry.get()) if (tel_entry.get() != None and tel_entry.get() != '') else ''
            tc = int(tc_entry.get()) if (tc_entry.get() != None and tc_entry.get() != '') else ''

            
            adress = adress_entry_l.get()+", "+ adress_entry_t.get()+", "+adress_entry_c.get()
            complain = complain_entry.get(1.0,END)
            inspection = inspection_entry.get(1.0,END)
            plan = plan_entry.get(1.0,END)
            
        except ValueError:
            error_text.config(text='Hata: Lütfen Yaş, TC Kimlik Numarası ve Telefon Numarası alanlarının\n numerik bir girdi olduğuna emin olunuz.\n Bir değer girmeyecekseniz lütfen boş bırakınız')
            return
        
        op_id=values[0]
        
        row=createElement(
        op_id,
        date,
        firstname,
        lastname,
        age,
        sex,
        tel,
        tc,
        adress,
        complain,
        inspection,
        plan
        )
        
        editElement(op_id, row)
        tree.item(selected, values=(row[0]))
        edit_patient.destroy()


    reg_button= Button(edit_patient, text="Hastayı Kaydet", command = edit_patient_info, bg='orange', fg='white', width="30")
    reg_button.place(x="155", y="700", height="30")
    
    error_text = Label(edit_patient, text='')
    error_text.place(x="80", y="750")
    
    edit_patient.mainloop()

# 3.3 Create backup screen
def backup_window():
    now = datetime.now()
    save_backup = Tk()
    save_backup.resizable(height = False, width = False)
    save_backup.geometry("520x80")
    save_backup.title("Yedek Oluştur")
    
    backup_name = StringVar()
    backup_name_entry = Entry(save_backup, textvariable= backup_name,width="80")
    backup_name_entry.pack(pady=10,padx=5)
    
    date = now.strftime("%m-%d-%Y-%H-%M-%S")
    filename= ("patient-data-backup-"+date)
    backup_name_entry.insert(0, filename)
    
    def saveBackup():
        global file_data
        file_data.to_csv('./'+backup_name_entry.get()+'.csv') 
        save_backup.destroy()
    
    button_save = Button(save_backup,text="Kaydet", command =lambda:saveBackup(), bg='orange', fg='white', width="5")
    button_save.pack()
    save_backup.mainloop()
# 4th: Create PDF Files:
def createPDF(data):
    
    #content
    filename = 'patient_pdf.pdf'
    documentTitle = 'patient information'
    title = 'Doktor İzlem Formu'
    
    dateLabel = 'Tarih:'
    nameLabel = 'Hasta Adı Soyadı:'
    ageLabel = 'Yaşı:'
    sexLabel = 'Cinsiyeti:'
    telLabel = 'Cep Tel:'
    idLabel = 'TC:'
    locationLabel = 'Adres:'
    complainLabel = 'YAKINMA:'
    inspLabel  = 'FİZİK MUAYENE:'
    planLabel = 'PLAN:'
    
    
    date = str(data[1])
    name = str(data[2]) + ' ' + str(data[3])
    age = str(data[4])
    sex = str(data[5])
    tel = str(data[6])
    idno = str(data[7])
    location = str(data[8])
    complain = str(data[9]).split("\n")
    insp = str(data[10]).split("\n")
    plan = str(data[11]).split("\n")
    
    
    # 0) Create Document :
    from reportlab.pdfgen import canvas

    
    pdf = canvas.Canvas(filename)
    pdf.setTitle(documentTitle)
    
    # 1) Title :: Set fonts
    
        
    
    from reportlab.pdfbase.ttfonts import TTFont
    from reportlab.pdfbase import pdfmetrics
    
    pdfmetrics.registerFont(TTFont("tr-font",'Arial.ttf'))
    pdf.setFont("tr-font",36)
    
    pdf.drawCentredString(290, 770, title)
    
    



    
    
    # 2) Labels
    
    pdf.setFont("tr-font",12)
    pdf.drawString(190, 730, dateLabel)
    pdf.drawString(190, 700, nameLabel)
    pdf.drawString(190, 670, ageLabel)
    pdf.drawString(350, 670, sexLabel)
    pdf.drawString(190, 640, telLabel)
    pdf.drawString(350, 640, idLabel)
    pdf.drawString(190, 610, locationLabel)
    pdf.drawString(190, 550, complainLabel)
    pdf.drawString(190, 470, inspLabel)
    pdf.drawString(190, 390, planLabel)
    
    
    # 3) Contents
    
    pdf.drawString(230, 730, date)
    pdf.drawString(303, 700, name)
    pdf.drawString(230, 670, age)
    pdf.drawString(403, 670, sex)
    pdf.drawString(238, 640, tel)
    pdf.drawString(370, 640, idno)
    pdf.drawString(227, 610, location)
    
    pdf.drawString(250, 550, complain[0])
    for i in range(1,4):
        if(i < len(complain)):
            pdf.drawString(250, 550-(i*15), complain[i])
        
    pdf.drawString(290, 470, insp[0])
    for i in range(1,4):
        if(i < len(insp)):
            pdf.drawString(290, 470-(i*15), insp[i])
        
    pdf.drawString(227, 390, plan[0])
    for i in range(1,4):
        if(i < len(plan)):
            pdf.drawString(227, 390-(i*15), plan[i])
    

        

    
    # F) Save The File
    import webbrowser
    pdf.save()
    webbrowser.open_new(filename)
# 5th Create table window
def table_window():
    root=Tk();
    root.title("Main")
    root.geometry("1250x500")

    #Define tree object
    datatree= ttk.Treeview(root)
    
    #Define columns from pandas dataframe
    datatree['columns'] = data_columns
    datatree['show'] = 'headings'
    
    #loop trough data tree and add headers
    for column in datatree["column"]:
        datatree.heading(column, text=column)
    
    refreshDisplay()
    updateTree(datatree)

    
    #Formate columns
    datatree.column("#0",width=25,minwidth=5)
    datatree.column("op id",width=40,minwidth=5)
    datatree.column("Tarih",width=100,minwidth=5)
    datatree.column("Ad",width=60,minwidth=5)
    datatree.column("Soyad",width=60,minwidth=5)
    datatree.column("Yaş",width=40,minwidth=5)
    datatree.column("Cinsiyet",width=60,minwidth=5)
    datatree.column("Cep Telefonu",width=120,minwidth=5)
    datatree.column("TC Kimlik No",width=120,minwidth=5)
    datatree.column("Adres",width=150,minwidth=5)
    datatree.column("Yakınma",width=180,minwidth=5)
    datatree.column("Fizik Muayene",width=150,minwidth=5)
    datatree.column("Plan",width=120,minwidth=5)
    
    
    
    
    datatree.pack(padx=20,pady=20)

            
    buttonFrame = Frame(root)
    
    button_add = Button(buttonFrame,text="Yeni Oluştur", command =lambda:patient_add_form_window(datatree), bg='orange', fg='white', width="15")
    button_add.pack(pady=20,padx=(0,30), side="left")
    button_delete = Button(buttonFrame,text="Seçileni Sil", command =lambda:deleteSelection(datatree, datatree.selection()[0]), bg='orange', fg='white', width="15")
    button_delete.pack(pady=20,padx=(0,30), side="left")
    button_edit = Button(buttonFrame,text="Seçileni Düzenle", command =lambda:patient_edit_form_window(datatree, datatree.selection()[0]), bg='orange', fg='white', width="15")
    button_edit.pack(pady=20,padx=(0,30), side="left")
    buttonFrame.pack()

    
    
    searchFrame = Frame(root)    
    search = StringVar()
    search_entry = Entry(searchFrame, textvariable= search)
    search_entry.pack(pady=20, side="left")
    button_search = Button(searchFrame,text="Ara", command =lambda:searchElement(search_entry.get(), datatree), bg='orange', fg='white', width="5")
    button_search.pack(pady=20,padx=(5,0), side="left")
    searchFrame.pack()
    
    saveFrame = Frame(root)
    button_save = Button(saveFrame, text="Yedek Kaydet", command =backup_window, bg='orange', fg='white')
    button_save.pack()
    saveFrame.pack()
    
    viewFrame = Frame(root)
    button_view = Button(saveFrame, text="PDF Oluştur", command =lambda: createPDF(datatree.item(datatree.selection()[0])['values']), bg='orange', fg='white')
    button_view.pack()
    viewFrame.pack()

    root.mainloop()
table_window()
```
